```python
# 2. Load the .pt dataset
import torch
import os

base = "/content/drive/MyDrive/ML Datasets"  # change if needed
path = os.path.join(base, "osd569_spike_dataset_10x.pt")

data = torch.load(path)

X_spikes = data["X_spikes"]   # (N, T, F)
y = data["y"]                 # (N,)
T = data["T"]
feature_cols = data["feature_cols"]

print("X_spikes shape:", X_spikes.shape)
print("y shape:", y.shape)
print("T:", T)
print("Number of features:", X_spikes.shape[-1])
print("Unique labels:", torch.unique(y, return_counts=True))
```

```python
from torch.utils.data import Dataset, DataLoader, random_split

class OSD569SpikeDataset(Dataset):
    def __init__(self, X_spikes, y):
        self.X = X_spikes.float()   # (N, T, F)
        self.y = y.long()           # (N,)

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

dataset = OSD569SpikeDataset(X_spikes, y)
print("Total samples:", len(dataset))
```

```python
# Train/val/test split: 60/20/20
torch.manual_seed(42)

N = len(dataset)
n_train = int(0.6 * N)
n_val   = int(0.2 * N)
n_test  = N - n_train - n_val

train_set, val_set, test_set = random_split(dataset, [n_train, n_val, n_test])
print(len(train_set), len(val_set), len(test_set))
```

```python
# DataLoaders
batch_size = 32

train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True,  drop_last=False)
val_loader   = DataLoader(val_set,   batch_size=batch_size, shuffle=False, drop_last=False)
test_loader  = DataLoader(test_set,  batch_size=batch_size, shuffle=False, drop_last=False)

xb, yb = next(iter(train_loader))
print("Batch X shape:", xb.shape)  # (B, T, F)
print("Batch y shape:", yb.shape)  # (B,)
```

```python
import torch.nn as nn
import torch.nn.functional as F
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

F_in = X_spikes.shape[2]
n_out = 2  # binary labels: 0, 1
```

```python
# Simple leaky integrate-and-fire layer
class SimpleLIF(nn.Module):
    def __init__(self, beta=0.9, threshold=1.0):
        super().__init__()
        self.beta = beta
        self.threshold = threshold

    def forward(self, x):

        batch_size, T_steps, H = x.shape
        mem = torch.zeros(batch_size, H, device=x.device)
        spk_rec = []

        for t in range(T_steps):
            mem = self.beta * mem + x[:, t, :]
            spk = (mem >= self.threshold).float()
            mem = mem * (1.0 - spk)
            spk_rec.append(spk)

        spk_rec = torch.stack(spk_rec, dim=1)
        return spk_rec
```

SNN model (2 hidden linear layers + LIF)

```python

class OSD569SNN(nn.Module):
    def __init__(self, F_in, n_hidden1=128, n_hidden2=64, n_out=2,
                 beta=0.9, threshold=1.0):
        super().__init__()
        self.fc1 = nn.Linear(F_in, n_hidden1)
        self.fc2 = nn.Linear(n_hidden1, n_hidden2)
        self.lif = SimpleLIF(beta=beta, threshold=threshold)
        self.fc_out = nn.Linear(n_hidden2, n_out)

    def forward(self, x):
        h1 = torch.relu(self.fc1(x))
        h2 = torch.relu(self.fc2(h1))
        spk = self.lif(h2)
        spk_mean = spk.mean(dim=1)
        out = self.fc_out(spk_mean)
        return out
```

```python
def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss = 0.0
    all_preds = []
    all_targets = []

    for xb, yb in loader:
        xb = xb.to(device)
        yb = yb.to(device)

        optimizer.zero_grad()
        outputs = model(xb)
        loss = criterion(outputs, yb)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * xb.size(0)

        preds = outputs.argmax(dim=1)
        all_preds.append(preds.detach().cpu())
        all_targets.append(yb.detach().cpu())

    all_preds = torch.cat(all_preds).numpy()
    all_targets = torch.cat(all_targets).numpy()

    acc = accuracy_score(all_targets, all_preds)
    return total_loss / len(loader.dataset), acc
```

```python
def evaluate(model, loader, criterion, compute_full_metrics=False):
    model.eval()
    total_loss = 0.0
    all_preds = []
    all_targets = []

    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device)
            yb = yb.to(device)

            outputs = model(xb)
            loss = criterion(outputs, yb)
            total_loss += loss.item() * xb.size(0)

            preds = outputs.argmax(dim=1)
            all_preds.append(preds.cpu())
            all_targets.append(yb.cpu())

    all_preds = torch.cat(all_preds).numpy()
    all_targets = torch.cat(all_targets).numpy()

    acc = accuracy_score(all_targets, all_preds)

    if not compute_full_metrics:
        return total_loss / len(loader.dataset), acc

    precision = precision_score(all_targets, all_preds, zero_division=0)
    recall = recall_score(all_targets, all_preds, zero_division=0)
    f1 = f1_score(all_targets, all_preds, zero_division=0)

    return (total_loss / len(loader.dataset),
            acc, precision, recall, f1)
```

```python
model = OSD569SNN(F_in=F_in,
                  n_hidden1=128,
                  n_hidden2=64,
                  n_out=2,
                  beta=0.9,
                  threshold=1.0).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=3e-3, weight_decay=1e-5)
num_epochs = 150

best_val_acc = 0.0
best_state = None

for epoch in range(1, num_epochs + 1):
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion)
    val_loss, val_acc = evaluate(model, val_loader, criterion)

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_state = model.state_dict()

    if epoch % 10 == 0:
        print(f"Epoch {epoch:03d} | "
              f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
              f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")

if best_state is not None:
    model.load_state_dict(best_state)
```

```python
import numpy as np

def subset_to_numpy(subset):
    X_list = []
    y_list = []
    for X_i, y_i in subset:
        X_list.append(X_i.numpy())
        y_list.append(y_i.numpy())
    X_arr = np.stack(X_list, axis=0)
    y_arr = np.array(y_list)
    return X_arr, y_arr

X_train_seq, y_train_np = subset_to_numpy(train_set)
X_val_seq,   y_val_np   = subset_to_numpy(val_set)
X_test_seq,  y_test_np  = subset_to_numpy(test_set)

print("Train seq shape:", X_train_seq.shape)
print("Val seq shape  :", X_val_seq.shape)
print("Test seq shape :", X_test_seq.shape)
```

```python
def seq_to_features(X_seq):
    mean_feat = X_seq.mean(axis=1)
    var_feat  = X_seq.var(axis=1)
    X_flat = np.concatenate([mean_feat, var_feat], axis=1)
    return X_flat

X_train_flat = seq_to_features(X_train_seq)
X_val_flat   = seq_to_features(X_val_seq)
X_test_flat  = seq_to_features(X_test_seq)

print("Flat train shape:", X_train_flat.shape)
print("Flat val shape  :", X_val_flat.shape)
print("Flat test shape :", X_test_flat.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

rf = RandomForestClassifier(
    n_estimators=50,
    max_depth=2,
    random_state=42
)

rf.fit(X_train_flat, y_train_np)

y_test_pred = rf.predict(X_test_flat)

rf_acc = accuracy_score(y_test_np, y_test_pred)
rf_prec = precision_score(y_test_np, y_test_pred, zero_division=0)
rf_rec  = recall_score(y_test_np, y_test_pred, zero_division=0)
rf_f1   = f1_score(y_test_np, y_test_pred, zero_division=0)

print(f"RF Test Acc : {rf_acc:.4f}")
print(f"RF Precision: {rf_prec:.4f}")
print(f"RF Recall   : {rf_rec:.4f}")
print(f"RF F1       : {rf_f1:.4f}")
```
